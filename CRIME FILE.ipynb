# Visualizations

### Top 10 States with Highest Rape Incidents

```python
state=crime.groupby('STATE/UT')['Rape'].sum().sort_values(ascending=False).head(10)
state
```

```python
state.plot(kind='bar', color='purple', figsize=(8,6))
plt.title('Top 10 States with Highest Rape Incidents')
plt.xlabel('State/UT')
plt.ylabel('Number of Rape Incident')
plt.show()
```

- **Madhya Pradesh has consistently reported the highest number of rape incidents among all states, raising serious concerns about the safety and security of women in the region.** 
- **This alarming statistic suggests that there are systemic issues contributing to violence against women.**

### Top 10 States with Highest Kidnapping and Abduction Incidents

```python
state_kidnapping_totals = crime.groupby('STATE/UT')['Kidnapping and Abduction'].sum().sort_values(ascending=False)
state_kidnapping_totals
```

```python
state_kidnapping_totals.head(10).plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8), startangle=90)
plt.title('Top 10 States with Highest Kidnapping and Abduction Incidents')
plt.ylabel('')  
plt.show()
```

- **Uttar Pradesh has the highest number of kidnapping and abduction incidents in India.**


# Top 10 States with Highest Dowry Deaths

```python
state_dowry_deaths = crime.groupby('STATE/UT')['Dowry Deaths'].sum().sort_values(ascending=False).head(10)
state_dowry_deaths
```

```python
state_dowry_deaths.head(10).plot(kind='bar', color='red', figsize=(10,6))
#plt.title('Top 10 States with Highest Dowry Deaths')
plt.xlabel('State/UT')
plt.ylabel('Number of Dowry Deaths')
plt.xticks(rotation=45)
plt.show()
```

- **Uttar Pradesh has the highest number of dowry deaths in India, indicating a serious violation of women's rights and a critical public safety issue that demands immediate attention**

# Top 10 States with Cruelty by Husband or Relatives Incidents

```python
state_cruelty = crime.groupby('STATE/UT')['Cruelty by Husband or his Relatives'].sum().sort_values(ascending=False)
state_cruelty
```

```python
state_cruelty.head(10).plot(kind='bar', color='brown', figsize=(10,6))
#plt.title('Top 10 States with Cruelty by Husband or Relatives Incidents')
plt.xlabel('State/UT')
plt.ylabel('Number of Cruelty Incidents')
plt.xticks(rotation=45)
plt.show()
```

- **West Bengal has the highest number of cruelty incidents by husbands or relatives in India, indicating a pressing concern for women's safety and rights within domestic settings.**
